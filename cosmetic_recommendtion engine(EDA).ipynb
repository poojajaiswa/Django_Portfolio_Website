```python
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(10,5))
sb.countplot(x='Rank',data=df)
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,5))
sb.catplot(x="Label",y="Price", data=df)
plt.title('Price per catagory')
```

```python
plt.figure(figsize=(12,7))
sb.boxplot(x="Rank",y="Price", data=df)
plt.title('Price by rating')
```

```python
df.groupby(by="Brand",axis=0).count()
```

```python
sb.heatmap(data=df.groupby(by="Brand",axis=0).count())
```

```python
plt.figure(figsize=(10,5))
sb.scatterplot(data=df,x='Rank',y='Price',hue='Label')
plt.title('Rating vs price')
```
