```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# assume you have a list of text documen

# create a TfidfVectorizer object
vectorizer = TfidfVectorizer()

# fit the vectorizer to the documents and transform them into a matrix
tfidf_matrix = vectorizer.fit_transform(df['tags'])

# compute the cosine similarity matrix
similarity_matrix = cosine_similarity(tfidf_matrix,tfidf_matrix)

print(similarity_matrix)
```

```python
similarity_df = pd.DataFrame(similarity_matrix, index=df['Name'], columns=df)

def recommend(item_id, num_recs=5):
    # get the similarity scores for the given item
    scores = similarity_df.loc[item_id]
    
    # sort the scores in descending order
    scores = scores.sort_values(ascending=False)
    
    # get the top N recommended items
    recs = scores.index[:num_recs]
    
    return recs

# example usage:
item_id = "The Moisturizing Soft Cream"
recommend(item_id, num_recs=5)
```

```python
def give_rec(title):
    vectorizer = TfidfVectorizer()

# fit the vectorizer to the documents and transform them into a matrix
    tfidf_matrix = vectorizer.fit_transform(df['tags'])

# Compute the sigmoid kernel
    sig = sigmoid_kernel(tfidf_matrix, tfidf_matrix)
    indices = pd.Series(df.index, index=df['Name']).drop_duplicates()
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return df.iloc[indices].sort_values(by='Rank',ascending=False)
```
